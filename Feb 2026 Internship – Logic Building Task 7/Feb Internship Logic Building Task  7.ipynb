# Problem 1: Smart Light Controller

```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status ="OFF"

    def switch(self,action):
        self.status = action
        print(f"{self.name} is {self.status}")

light = SmartLight("Bedroomm Light")
light.switch("ON")
```

# Problem 2: Employee ID Card System

```python
class Employee:
    def __init__(self, name,emp_id,department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display(self):
        print("Employee ID card")
        print("Name:",self.name)
        print("ID :",self.emp_id)
        print("Department:",self.department)

emp = Employee("Rahul","EMP20","IT")
emp.display()
```

# Problem 3: Mobile Contact Record

```python
class Contact:
    def __init__(self,name,number):
        self.name = name
        self.number = number

    def Save(self):
        print("Contact Saved")
        print("Name:",self.name)
        print("Phone:",self.number)

contact = Contact("Anita",9876543210)
contact.Save()
```

# Problem 4: Product Price Tag Generator

```python
class Product:
    def __init__(self,name,price):
        self.name = name
        self.price = price
        
    def price_tag(self):
        print("Product:",self.name)
        print(f"Price: Rs{self.price}")

product = Product("Headphones",2499)
product.price_tag()
```

# Problem 5: Movie Rating Display System

```python
class Movie:
    def __init__(self,name,rating):
        self.name = name
        self.rating = rating
    def display(self):
        print("Movie :",self.name)
        print(f"Rating : {self.rating}/5")

movie = Movie("Inception",4.8)
movie.display()
```

# Problem 6: Delivery Address Manager

```python
class Delivery_Details:
    def __init__(self,name,address):
        self.name = name
        self.address = address

    def Details(self):
        print("Delivery Details")
        print("Customer:",self.name)
        print("Address:",self.address)

details = Delivery_Details("Suman","Hydrabad")
details.Details()
```
