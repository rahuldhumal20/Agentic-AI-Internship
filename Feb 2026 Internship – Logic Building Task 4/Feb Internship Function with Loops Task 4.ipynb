```python
# Problem 1: Mobile Recharge Validation System
```

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    
    while True:                                               #User should be allowed to retry if invalid
        amount = int(input("Enter recharge amount : Rs"))

        if amount < 50:                                        #Recharge amount must be ≥ ₹50
            print("Recharge amount must be at least Rs 50")

        elif amount not in valid_plans:
            print("Invalid Plans . Available plans : 199,299,399,599")
        else:
            print("Recharge Successful")
            break
validate_recharge()       
```

```python
# Problem 2: Inventory Reorder Alert System
```

```python
def check_inventory(products):
    for product,stock in products.items():
        if stock < 15:                               #Print Reorder Alert if stock < 15
            print(f"{product}:Recoder Alert")
        else:
            print(f"{product}: Stock OK")

inventory = {
    "Laptop" : 10,
    "Mouse" : 25,
    "Keyboard":12,
    "Monitor": 20
}

check_inventory(inventory)
        
```

```python
# Problem 3: Student Result Processing System
```

```python
def process_result(marks):    
    total = 0
    
    for mark in marks:
        total += mark

    average = total / len(marks)            #Calculates average marks

    print("Average Marks:",average)

    if average >= 50:                      #Prints Pass if average ≥ 50 else Fail
        return "Pass"
    else:
        return "Fail"
        
marks = [60,75,70,40]
result = process_result(marks)

print("Result:", result)
```

```python
# Problem 4: Cab Fare Estimator with Retry Option
```

```python
def calculate_fare(distance, peak=False):
    base_fare = 50                          #Base fare = ₹50
    
    fare = base_fare + (12 * distance)       

    if peak:                                #Peak hour → 25% extra
        fare += fare * 0.25
        
    return fare

while True :
    distance = float(input("Enter distance in km :"))
    peak_input = input("Is it peak hous ? (yes/ no) :").lower()

    peak = True if peak_input == "yes" else False

    total_fare = calculate_fare(distance, peak)
    
    print(f"Total Fare: Rs{total_fare}")

    retry = input("Do you want to calculate ? (yes/ no): " ).lower()          # Allow retry using while loop
    if retry != "yes":
        break
```

```python
# Problem 5: Employee Attendance Eligibility Checker
```

```python
def check_attendence(attendence_list):
    precent_days = 0

    for day in attendence_list:
        if day == "P":
            precent_days += 1

    percentage = (precent_days / len(attendence_list)) * 100           #Calculates attendance percentage
    print("Attendence Percentage :",percentage)

    if percentage >= 75:                #Returns Eligible if ≥ 75% else Not Eligible
        return "Eligible"
    else:
        return "Not Elligible"

attendence = ["P" ,"P", "A", "A" ,"P","P", "P"]
status = check_attendence(attendence)
print("Eligibility Status:",status)

    
```

```python
# Problem 6: Password Strength Checker
```

```python
def check_password(password):
    if len(password) < 8 :
        return " Weak Password: Minimum length shold be 8"

    has_digit = False
    has_special = False
    special_chars = "@#$"

    for char in password:
        if char.isdigit():                   #Must contain at least one digit
            has_digit = True
        if char in special_chars:            #Must contain at least one special character (@#$)
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password Must contains digit and special character(@#$)"
    
password = input("Enter Password")
print(check_password(password))
            
```
