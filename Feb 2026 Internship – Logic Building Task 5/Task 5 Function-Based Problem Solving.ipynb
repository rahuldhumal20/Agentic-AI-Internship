```python
#Problem 1: Social Media – Post Engagement Analyzer
```

```python
def analyze_engaement(likes):
    total_likes = 0
# Calculate total likes

    for like in likes:
        total_likes += like

    status ="Viral Post" if total_likes >= 1000 else "Normal Enagagement"

    print("Total Likes : ",total_likes)
    print("Post Status : ", status )

analyze_engaement([200,33,55,888,155])
    
```

```python
#Problem 3: Agriculture – Rainfall Adequacy Checker
```

```python
def reainfall_checker(rainfall,required):

# Calculate average rainfall 
    avg = sum(rainfall) / len(rainfall)

    status = "Adequate Rainfall " if avg >= required else "Inadequate Rainfall"

    print("Average Rainfall:",avg)
    print("Rainfall status:",status)

reainfall_checker([72,50,68,75],70)
```

```python
# Problem 4: Social Media – Duplicate Account Detection
```

```python
def check_duplicates(users):
    if len(users) != len(set(users)):
        print("Duplicate Account Found : Yes ")
    else:
        print("Duplicate accounts Found : No")

check_duplicates(["rahul","amit","sumit","rahul"])
```

```python
#Problem 7: System Monitoring – Application Health Checker
```

```python
def application_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status ="Critical Issues"

    print("Error Count :",errors)
    print("System Status:",status)

application_health(6)
```

```python
#Problem 9: E-Learning – Student Attendance Eligibility System
```

```python
def attendence_check(attendence):
    present = attendence.count("P")
    
# calculate  Percentage
    
    percentage = (present / len(attendence)) * 100
    
# status

    status = "Eligible" if percentage >= 75 else "Not Eligible"

    print("Attendence Percentage :",percentage)
    print("Exam Eligibility :",status)

attendence_check(["P","P","A","P","A","P","P","P","P"])
    
```

```python
#Problem 10: Smart Electricity Bill Analyzer
```

```python
def electricity_bill(units):
    bill = 0
    
#Billing Rules    
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = units * 5
    else :
        bill = units * 7

# Usage Classification
    if bill < 500:
        usage ="Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage ="High Usage"

    print("Total Bill:",bill)
    print("Usage Level:",usage)

electricity_bill(155)

```
