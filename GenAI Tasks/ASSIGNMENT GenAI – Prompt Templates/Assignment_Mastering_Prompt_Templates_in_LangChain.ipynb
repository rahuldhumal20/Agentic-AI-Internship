# Task 5: Input Validation Layer

```python
def validate_inputes(audience, tone):
  valid_audience = ["beginner","intermidiate","expert"]
  valid_tone = ["formal","causal","fun"]

  if audience not in valid_audience:
    audience = "beginner"

  if tone not in valid_tone:
    tone = "formal"

  return audience, tone
```

# Task 6: Prompt Generator App

```python
def genrate_prompt(topic, audience,tone,style):
  audeince, tone = validate_inputes(audience,tone)

  templates = {
      "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
      "interview": "Ask interviw quetions about {topic} for audience in a {tone} tone",
      "storytelling": "Explain {topic} as a story for {audience} in a {tone} tone"
  }

  prompt = PromptTemplate(
      input_variables=["topic","audience", "tone"],
      template=templates[style]
  )

  result = prompt.format(
      topic=topic,
      audience=audience,
      tone=tone
  )

  print("\nGenrated Prompt:")
  return result

# Example

genrate_prompt(
   topic= "Neural Networks",
   audience = "beginner",
   tone = "fun",
   style = "storytelling"
  )
```
