```python
import os
from getpass import getpass
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.tools import tool

os.environ["OPENAI_API_KEY"] = getpass("Enter your Groq API key: ")
os.environ["OPENAI_BASE_URL"] = "https://api.groq.com/openai/v1"
```

```python
# Single LLM instance reused across all sections
llm = ChatOpenAI(model="llama-3.3-70b-versatile")
```

# 1. Basic LLM Call

```python

print("[1] Basic LLM Call:")
print(llm.invoke("What is Machine Learning?").content)
```

# 2. Prompt Template

```python

print("\n[2] Prompt Template:")
prompt = PromptTemplate.from_template("Explain {topic} in simple terms.")
print(prompt.format(topic="Deep Learning"))
```

# 3. LCEL Chain

```python

print("\n[3] Chain Output:")
chain = prompt | llm
print(chain.invoke({"topic": "Neural Networks"}).content)
```

# 4. Memory

```python

print("\n[4] Memory Conversation:")
chat_history = []

def chat(user_input):
    chat_history.append(f"User: {user_input}")
    reply = llm.invoke("\n".join(chat_history) + "\nAssistant:").content
    chat_history.append(f"Assistant: {reply}")
    return reply

print("Bot:", chat("Hi, my name is Rahul"))
print("Bot:", chat("I like cars"))
print("Bot:", chat("What do I like?"))
```

# 5. Tool (Calculator)

```python

print("\n[5] Agent Tool Output:")

@tool
def calculator(expression: str) -> str:
    """Evaluates a math expression like '25 * 4'."""
    try:
        return str(eval(expression))
    except:
        return "Invalid expression."

print("25 * 4 =", calculator.invoke("25 * 4"))
```


# 6. Modular Pattern

```python

print("\n[6] Modular Output:")

def create_llm(model="llama-3.3-70b-versatile"):
    """Factory function — returns a configured ChatOpenAI instance."""
    return ChatOpenAI(model=model)

def explain_topic(topic: str) -> str:
    """Generates a plain-English explanation for any given topic."""
    chain = PromptTemplate.from_template("Explain {topic} simply.") | create_llm()
    return chain.invoke({"topic": topic}).content

print(explain_topic("Artificial Intelligence"))
```
