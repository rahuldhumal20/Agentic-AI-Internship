1. What is the difference between "Love" and "love" in NLP?

Ans: In NLP ,text data is often case sensetive , which means that words like "Love" and love" are treted as two different tokens.
     This can also lead to redudency and inconsistency in dataset.
     To avoid this issue, a common preprocessing step called case normalization is applied, where all text is converted to lowercase.
     This ensures that words with the same meaning are treated uniformly.

     Example : 
         "Love you" → ["Love", "you"]
         "love you" → ["love", "you"]
        
     After normalization:
        Both become → ["love", "you"]

4. Explain the difference between stemming and lemmatization.
Ans: Both techniques are used to reduce words to their base form.

        Stemming
            Removes suffixes using simple rules
            May produce incorrect or non-meaningful words
    
        Example:
            "running" → "run"
            "studies" → "studi"
    
        Lemmatization
            Uses vocabulary and grammar rules
            Produces meaningful base words (lemma)
    
        Example:
            "running" → "run"
            "better" → "good"




Task 2:Build Advanced Preprocessing Function

```python
import re

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    # remove URls
    text = re.sub(r'http\S+|www\S+', '', text)

    # remove emails
    text = re.sub(r'\S+@\S+', '', text)

    #remove numbers
    text = re.sub(r'\d+', '', text)

    #convert to lowercase
    text = text.lower()

    #handle repeted characters (soooo -> so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    #remove special characters
    text = re.sub(r'[^\w\s]', '',text)

    #remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    #tokenization
    tokens = text.split()

    #remove short tokes (length <= 2 except 'no, 'not')
    tokens = [word for word in tokens if len(word) > 2 or word in ['no','not']]

    clean_sentence = " ".join(tokens)

    return tokens,clean_sentence

# Task 4 Token Analysis

def token_analysis(tokens):
    total = len(tokens)
    unique = len(set(tokens))
    avg_length = sum(len(t) for t in tokens) / total if total > 0 else 0
    return total, unique, avg_length
```

Task 3: Stress Testing , Task 4: Token Analytics

```python
text_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for sentence in text_sentences:
    tokens, clean = preprocess_text(sentence)

    total,unique, avg = token_analysis(tokens)
    
    print("\nOriginal Text:", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean)

    print("Total tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", round(avg, 2))
```

Task 5 : Frequency Analysis

```python
from collections import Counter

all_tokens = []

for sentence in text_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 frequent:", counter.most_common(10))
print("Least 5 frequent:", counter.most_common()[-5:])
```

Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
result = full_pipeline(text_sentences)

print("\n_____FULL PIPELINE OUTPUT_____")
    
print("\nAll Tokens:")
print(result["tokens"])
    
print("\nClean Sentences:")
for sentence in result["clean_sentences"]:
    print(sentence)
```
