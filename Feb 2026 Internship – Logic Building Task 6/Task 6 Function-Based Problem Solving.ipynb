# Problem Statement 1:Smart Parking Lot Management System

```python
def parking_system(capacity, logs):
    parked =0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        peak = max(peak,parked)
# status 
    status = "Full" if parked >= capacity else "Available"

    print("Currently Parked Vehicles:",parked)
    print("Parking Status:",status)

parking_system(50,["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

# Problem Statement 2: Online Food Delivery Time Estimator

```python
def delivery_time(distance,traffic,weather):
    # base 5 min per km
    time = distance * 5

    if traffic =="High":
        time += 10
    elif traffic == "Medium":
        time += 5

    if weather == "Rainy":
        time += 10
    elif weather == "Stromy":
        time += 20

    print("Estimated Delivery Time :",time , "Minutes")

delivery_time(8 ,"High","Rainy")
        
```

#  Problem Statement 3: Movie Theatre Seat Occupancy Analyzer

```python
def theatre_occupancy(total_seats, booked):
    occupied = len(booked)
#calculate percentage 
    
    percentage = (occupied / total_seats) * 100

# status     
    if percentage == 100:
        status = "Housefull"
    elif percentage >= 75 :
        status = "Almost full"
    else :
        status = "Available"

    print("Occupancy :",f"{percentage:.0f}%")
    print("Show Status:",status)

theatre_occupancy(200,[1]* 150)
```

# Problem Statement 4: Cloud Server Load Classification System

```python
def server_load(cpu_readings):

#calculate average 
    average = sum(cpu_readings)/len(cpu_readings)
#status
    
    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else :
        status = "Critical"

    print("Average CPU Load:",f"{average:.0f}")
    print("Server Status:",status)
    
server_load([50,60,80,50])
```

#  Problem Statement 5 : Smart Classroom Resource Usage Monitor

```python
def resource_monitor(resources):
    overused= []
    
    for resource, hours in resources.items():
        if hours > 8:              # apply condition
            overused.append(resource)

    alert = "Yes" if overused else "No"

    #output 
    
    print("Overused Resources:", ", ".join(overused))
    print("Energy Alert : ",alert)

# call

resource_monitor(
{
 "Projector": 6,
 "AC": 9,
 "Lights": 4
})
```

# Problem Statement 6: Online Event Registration Capacity Controller

```python
def event_registration(capacity,registrations):
    confirmed = min(capacity, registrations)
    waitlisted = max(0, registrations - capacity)
    
#status
    status = "closed" if registrations >= capacity else "Open"

    print("Confiremed Reg. :",confirmed)
    print("Waitlisted Users:",waitlisted)
    print("Registration Status: ",status)

# call

event_registration(100,105)

```
