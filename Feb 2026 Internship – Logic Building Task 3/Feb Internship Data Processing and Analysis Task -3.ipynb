```python
#1. Employee Performance Bonus Eligibility
```

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
#Find Highest Score
highest_score = max(employees.values())

top_performances = []

for name, score in employees.items():
    if score == highest_score:
        top_performances.append(name)

print("Top Performers Eligible for Bonous :",", ".join(top_performances), f"(Score: {highest_score})")
```

```python
#2. Search Query Keyword Analysis
```

```python
import string 

query = "Buy mobile phone buy phone online"

#convert to lowercase
query = query.lower()

#Remove Duplicates
query = query.translate(str.maketrans('','',string.punctuation))

words = query.split()

frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1

result = {}
for word,count in frequency.items():
    if count > 1:
        result[word] = count

print(result)
```

```python
sensor_readings = [3,4,7,8,10,12,5]

valid_readings = []

for index, value in enumerate(sensor_readings):
    if value % 2 == 0:      #even readins are valid 
        valid_readings.append((index, value))

print("Valid sensor Readings(Hour , value):")
print(valid_readings)
```

```python
# 4. Email Domain Usage Analysis
```

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0)+ 1

total_users = len(emails)

print("Email Domain Usage Percentage:")

for domain , count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# 5. Sales Spike Detection
```

```python
sales = [1200,1500,900,2200,1400,3000]

average_sales = sum(sales) / len(sales)

threshold = average_sales * 1.3  # 30% above average

for day , value in enumerate(sales , start = 1):
    if value > threshold:
        print(f"Day {day}:{value}")
```

```python
# 6. Duplicate User ID Detection
```

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

frequency = {}

for user in user_ids:
    frequency[user] = frequency.get(user,0)+1

for user, count in frequency.items():
    if count > 1:
        print(f"{user} -> {count} times")
```
