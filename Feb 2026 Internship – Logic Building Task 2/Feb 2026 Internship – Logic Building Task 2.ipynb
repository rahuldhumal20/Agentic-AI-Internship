```python
# Problem 1: Unique Words in a Sentence
```

```python
sentence = "python is easy and python is powerful"

words = sentence.split()
unique_words = set(words)

print("Unique Words Count :", len(unique_words))
print("Unique words : ", unique_words)
```

```python
#Problem 2: Highest Salary from Employee Data
```

```python
employees = {
    "Ravi":75000,
    "Ankita":68000,
    "Kiran": 72000
}

highest_employee = max(employees, key=employees.get)

print("Highest Salary :", highest_employee ,"-",employees[highest_employee])
    
```

```python
#Problem 3: Find Maximum and Minimum Values
```

```python
numbers = [45,22,89,10,66]

max_value = numbers[0]
min_value = numbers[0]

for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

print("List:",numbers)
print("Max:", max_value)
print("Min:", min_value)
```

```python
#Problem 4: Count Products Above a Price Threshold
```

```python
prices = [ 450,1200,899,1500,300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Product above 1000:",count)
```

```python
#Problem 5: Calculate Attendance Percentage
```

```python
attendence = ["P","P","A","P","P"]

precent_count = attendence.count("P")
percentage = (precent_count /len(attendence)) * 100

print("Attendence Percentage :", percentage)
```

```python
#Problem 7: Count Character Frequency
```

```python
text= "Python"
char_freq = {}

for char in text:
    if char in char_freq:
        char_freq[char] += 1
    else:
        char_freq[char] = 1

print(char_freq)
```
